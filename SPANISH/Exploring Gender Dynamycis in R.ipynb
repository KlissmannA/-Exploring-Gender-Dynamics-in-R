```r
###Payment method By Gender ###

#MALE#
payment_male <- df%>%
  filter(Gender == "Male") %>%
  count(PaymentMethod) %>%
  mutate(percentage = n / sum(n) * 100)

#FEMALE#

payment_female <- df %>%
  filter(Gender == "Female") %>%
  count(PaymentMethod)  %>%
  mutate(percentage = n / sum(n) * 100)
```

```r
plot_male <- ggplot(payment_male, aes(x = "", y = n, fill = PaymentMethod)) +
  geom_bar(width = 1, stat = "identity") +
  coord_polar("y") +
  theme_void() +
  geom_text(aes(label = paste0(n, " - ", round(percentage, 1), "%")),
            position = position_stack(vjust = 0.5), color = "black", angle = 65, hjust = 0.5, size = 4.5) +  
  ggtitle("Payment Method - Male") +  
  theme(legend.position = "bottom",
        plot.title = element_text(hjust = 0.5))

plot_female <- ggplot(payment_female, aes(x = "", y = n, fill = PaymentMethod)) +
  geom_bar(width = 1, stat = "identity") +
  coord_polar("y") +
  theme_void() +
  geom_text(aes(label = paste0(n, " - ", round(percentage, 1), "%")),
            position = position_stack(vjust = 0.5), color = "black", angle = 65, hjust = 0.5, size = 4.5) +  
  ggtitle("Payment Method - Female") +  
  theme(legend.position = "bottom",
        plot.title = element_text(hjust = 0.5))

grid.arrange(plot_male, plot_female, nrow = 2)
```

```r
male_country_plot <- ggplot(male_country_count, aes(x = reorder(Country, -n), y = n, fill = Country)) +
  geom_bar(stat = "identity", color = "black", width = 0.6) +
  geom_text(aes(label = n), vjust = -0.5, color = "black", size = 5) + 
  labs(x = "Country", y = "Count") +
  ggtitle("Countries with Most Males") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1, size = 12),
    axis.title = element_text(size = 14),
    plot.title = element_text(size = 16, hjust = 0.5),
    legend.position = "none",
    panel.grid.major.x = element_blank(),
    panel.grid.minor = element_blank()
  ) +
  scale_fill_viridis_d()

female_country_plot <- ggplot(female_country_count, aes(x = reorder(Country, -n), y = n, fill = Country)) +
  geom_bar(stat = "identity", color = "black", width = 0.6) +
  geom_text(aes(label = n), vjust = -0.5, color = "black", size = 5) +
  labs(x = "Country", y = "Count") +
  ggtitle("Countries with Most Females") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1, size = 12),
    axis.title = element_text(size = 14),
    plot.title = element_text(size = 16, hjust = 0.5),
    legend.position = "none",
    panel.grid.major.x = element_blank(),
    panel.grid.minor = element_blank()
  ) +
  scale_fill_viridis_d()

grid.arrange(male_country_plot, female_country_plot, ncol = 2, padding = unit(1, "cm"))
```
